```python
#Numpyの基本的な関数や使い方についてまとめています
#備忘録的なものです
import numpy as np
```

```python
#-四則演算-
```

```python
a = 1
b = 2
#足し算
print(a + b)
#引き算
print(a - b)
#掛け算
print(a * b)
#指数(a^b)
print(a ** b)
#割り算
print(a / b)
#割り算(整数部分のみ)
print(a // b)
#割り算(余り部分のみ)
print(a % b)
```

```python
#-配列(Array)の作成-
```

```python
#リストの配列変換
```

```python
List = [1, 2, 3]
array = np.array(List)
print(array)
```

```python
#多次元配列の作成
```

```python
#リストから作成
List1 = [1, 2, 3]
List2 = [4, 5, 6]
array1 = np.array([List1, List2])
print(array1)
#arange関数で作成
array2 = np.arange(1,7).reshape(2,3)
print(array2)
```

```python
#要素が全て0の配列・多次元配列を作成
```

```python
#1次元
List1 = np.zeros(10)
print(List1)
#2次元
List2 = np.zeros((2,5))
print(List2)
```

```python
#要素が全て1の配列・多次元配列を作成
```

```python
#1次元
List1 = np.ones(10)
print(List1)
#2次元
List2 = np.ones((2,5))
print(List2)
```

```python
#単位行列を作成
```

```python
array = np.eye(5)
print(array)
```

```python
#連続した数値の配列を作成
```

```python
#0～9まで
array1 = np.arange(10)
print(array1)
#5～9まで
array2 = np.arange(5,10)
print(array2)
#0～100まで20間隔で
array3 = np.arange(0,101,10)
print(array3)
```

```python
#-インデックス操作
```

```python
#インデックスを指定して配列から値を抽出
```

```python
array = np.arange(1,10)
print(array)
#5を取得(5のインデックスは4)
print(array[4])
#8を取得(8のインデックスは7)
print(array[7])
```

```python
#インデックスを指定して行列から行を取得
```

```python
array = np.arange(9).reshape(3,3)
print(array)
#1行目を取得
print(array[0])
#3行目を取得
print(array[2])
```

```python
#インデックスを指定して行列から列を取得
```

```python
array = np.arange(9).reshape(3,3)
print(array)
#2列目を取得
print(array[:,1])
```

```python
#インデックスを指定して行列から値を取得
```

```python
array = np.arange(9).reshape(3,3)
print(array)
#2行目2列目の値を取得
print(array[1,1])
```

```python
#インデックスを指定して配列から特定範囲の値を抽出
```

```python
array = np.arange(1,10)
print(array)
#3～5を取得
print(array[2:5])
```

```python
#インデックスを指定して配列中の値を更新
```

```python
array = np.arange(1,10)
print(array)
#5を100に更新
array[4] = 100
print(array)
```

```python
#演算関数
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#加法
print(np.add(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#減法
print(np.subtract(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#乗法
print(np.multiply(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#除法
print(np.divide(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#べき乗
print(np.power(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
a2 = np.arange(9,18).reshape(3,3)
print(a1)
print(a2)
#行列積
print(np.dot(a1,a2))
```

```python
a1 = np.arange(9).reshape(3,3)
print(a1)
#平方根
print(np.sqrt(a1))
```

```python
a1 = np.arange(9).reshape(3,3)
print(a1)
#絶対値
print(np.abs(a1))
```
