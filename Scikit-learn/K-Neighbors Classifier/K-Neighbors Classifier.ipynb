```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
df.isnull().sum() #欠損値なしを確認
```

```python
LE = LabelEncoder()
df["Species"] = LE.fit_transform(df["Species"])
df.head()
```

```python
df.tail()
```

```python
X = df.drop("Species", axis=1)
y = df["Species"]
X.head()
```

```python
KN = KNeighborsClassifier(n_neighbors=5) #近傍点はとりあえず5とかで

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 0)

KN.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = KN.predict(X_test)
y_train_pred = KN.predict(X_train)

print("トレーニングデータの正答率：", accuracy_score(y_train, y_train_pred))
print("モデルの正答率：", accuracy_score(y_pred, y_test))
```
